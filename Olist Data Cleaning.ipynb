### Order-level Review Consolidation

Some orders contain multiple review records because items within the same order may be delivered at different times, or customers may update their feedback after the initial review.

In this project, the `orders` dataset is the primary source for delivery time and delivery performance analysis. However, this dataset stores delivery information at the order level, not at the individual item level. In other words, each `order_id` represents the overall order delivery outcome rather than separate delivery records for each item within that order.

Since the review dataset does not provide item-level review mapping, multiple reviews from the same order cannot be accurately linked to specific individual items or separate delivery events.

To keep the review data consistent with the order-level structure of the `orders` dataset, only the most recent review for each `order_id` is retained based on `review_answer_timestamp`.

The latest review score is used as the final measure of the customer's overall satisfaction with the complete order and its delivery performance.

```python
# Sort reviews by order_id and latest response timestamp
df_reviews_sorted = df_reviews_valid.sort_values(
    by=['order_id', 'review_answer_timestamp'],
    ascending=[True, False]
)

# Keep the latest review for each order
df_reviews_order_level = df_reviews_sorted.drop_duplicates(
    subset='order_id',
    keep='first'
)

df_reviews_order_level = df_reviews_order_level.reset_index(drop=True)
df_reviews_order_level.head()
```

### 7.4. Master Whole-order Table Construction

### Methodology: Master Whole-order Table Construction

This project is structured at the **order level** because the Olist dataset records delivery outcomes by `order_id`, not by individual item.

Although a single order may contain multiple items, products, or sellers, the delivery timestamps describe the overall delivery journey of the complete order. Therefore, `order_id` is used as the main unit of observation.

To build a consistent order-level dataset, the `orders` dataset is used as the central table because it contains the complete set of delivery timestamps and the unique `order_id`.

Before joining datasets, item-level information is first aggregated to one row per order. The `order_items` dataset is enriched with product information from the `products` dataset, then grouped by `order_id` to create order-level features such as:

- number of items;
- number of unique products;
- number of sellers;
- total product value;
- total freight cost;
- total product weight;
- total product volume.

These variables are not intended to identify which specific item caused a delivery delay. Instead, they describe the overall size, value, and operational complexity of each order.

The deduplicated `reviews` dataset is then joined to capture the final customer satisfaction outcome for each order, while customer-level information from the `customers` dataset is added to provide basic customer and geographic context.

After the master order-level table is created, each order is classified into one of two structural groups:

- **Homogeneous orders:** orders containing only one product type and one seller.
- **Complex orders:** orders containing multiple product types, multiple sellers, or both.

As a result, the final master table provides a unified and methodologically consistent foundation for analyzing delivery performance and customer satisfaction at the order level.

```python
# Create product-level features from the original products table
df_products_features = df_products.copy()

df_products_features['product_volume_cm3'] = (
    df_products_features['product_length_cm'] *
    df_products_features['product_height_cm'] *
    df_products_features['product_width_cm']
)

df_products_features.head(5)
```

```python
# Select product columns needed for order-level aggregation
product_cols = [
    'product_id',
    'product_category_name',
    'product_weight_g',
    'product_volume_cm3'
]

df_order_items_enriched = df_order_items.merge(
    df_products_features[product_cols],
    on='product_id',
    how='left'
)

df_order_items_enriched.head(5)
```

```python
# Check missing product information after merging
df_order_items_enriched[
    ['product_category_name', 'product_weight_g', 'product_volume_cm3']
].isna().sum()
```

```python
# Fill missing weight and volume using median values
df_order_items_enriched['product_weight_g'] = (
    df_order_items_enriched['product_weight_g']
    .fillna(df_order_items_enriched['product_weight_g'].median())
)

df_order_items_enriched['product_volume_cm3'] = (
    df_order_items_enriched['product_volume_cm3']
    .fillna(df_order_items_enriched['product_volume_cm3'].median())
)
```

```python
# Build order-level features from item-level data
df_order_features = df_order_items_enriched.groupby('order_id').agg(
    item_count=('order_item_id', 'count'),
    product_count=('product_id', 'nunique'),
    category_count=('product_category_name', 'nunique'),
    seller_count=('seller_id', 'nunique'),
    total_price=('price', 'sum'),
    total_freight=('freight_value', 'sum'),
    total_weight_g=('product_weight_g', 'sum'),
    total_volume_cm3=('product_volume_cm3', 'sum')
).reset_index()

df_order_features.head(5)
```

```python
# Create order structure flags
df_order_features['multi_seller_flag'] = (
    df_order_features['seller_count'] > 1
).astype(int)

df_order_features['multi_product_flag'] = (
    df_order_features['product_count'] > 1
).astype(int)

df_order_features['freight_to_price_ratio'] = np.where(
    df_order_features['total_price'] > 0,
    df_order_features['total_freight'] / df_order_features['total_price'],
    0
)
```

```python
# Classify order structure type
df_order_features['order_structure_type'] = np.where(
    (df_order_features['product_count'] == 1) &
    (df_order_features['seller_count'] == 1),
    'Homogeneous',
    'Complex'
)

df_order_features.head(5)
```

### Methodology: Geolocation Data Preparation and Integration

A key methodological challenge in this project is handling the `geolocation` dataset. In the raw Olist data, a single `geolocation_zip_code_prefix` can appear multiple times with slightly different latitude and longitude values. If this table is joined directly to the master dataset, it may create duplicate rows and many-to-many merge issues.

To avoid this problem, the geolocation data is first collapsed to one row per zip code prefix. For each `geolocation_zip_code_prefix`, the latitude and longitude are averaged, while the most frequent city and state values are retained. This creates a stable geographic reference table that can be safely joined to the order-level master dataset.

These geographic fields are treated as approximate location proxies, not exact address-level coordinates. This reflects a practical analytics approach: the goal is not to force artificial precision, but to build reliable geographic features that support meaningful operational analysis.

Another important decision involves seller geography. Since delivery outcomes in the Olist dataset are recorded once per completed order, a single seller location can only validly represent orders fulfilled by one seller. For multi-seller orders, assigning one arbitrary seller location would be misleading.

Therefore, seller geolocation is directly assigned only to single-seller orders. For complex or multi-seller orders, seller-related complexity is captured through variables such as `seller_count` and `multi_seller_flag`. This keeps the analysis consistent with the order-level structure while still preserving useful information about operational complexity.

```python
# Function to get the most frequent value
def first_mode(x):
    mode_value = x.mode()
    return mode_value.iloc[0] if not mode_value.empty else x.iloc[0]


# Collapse geolocation data to one row per zip code prefix
df_geo_clean = df_geo.groupby('geolocation_zip_code_prefix').agg(
    geolocation_lat=('geolocation_lat', 'mean'),
    geolocation_lng=('geolocation_lng', 'mean'),
    geolocation_city=('geolocation_city', first_mode),
    geolocation_state=('geolocation_state', first_mode)
).reset_index()

df_geo_clean.head(5)
```

### Time Logic Validation and Filtering

The final master table is validated to ensure that delivery-related timestamps follow a meaningful chronological order for delivery performance analysis.

Two types of timestamp inconsistencies are identified:

- `approval_after_carrier`: cases where `order_approved_at` is later than `order_delivered_carrier_date`.
- `carrier_after_customer`: cases where `order_delivered_carrier_date` is later than `order_delivered_customer_date`.

The `carrier_after_customer` cases are treated as hard inconsistencies because the carrier handoff should occur before customer delivery.

The `approval_after_carrier` cases may reflect payment approval delays, system recording issues, or specific payment workflows. However, these rows produce negative `seller_handling_time`, which conflicts with the feature definition used in this project.

Therefore, both types of inconsistent rows are removed from the analytical dataset to keep delivery time features logically valid and interpretable.

```python
# Remove rows with inconsistent delivery timestamp logic
df_master_orders_time_valid = df_master_orders_final[
    ~invalid_time_mask
].reset_index(drop=True)
```

```python
# Recheck timestamp inconsistencies after filtering
approval_after_carrier_check = (
    df_master_orders_time_valid['order_approved_at'] >
    df_master_orders_time_valid['order_delivered_carrier_date']
).sum()

carrier_after_customer_check = (
    df_master_orders_time_valid['order_delivered_carrier_date'] >
    df_master_orders_time_valid['order_delivered_customer_date']
).sum()

print("Approval after carrier rows:", approval_after_carrier_check)
print("Carrier after customer rows:", carrier_after_customer_check)
print("Rows after filtering:", len(df_master_orders_time_valid))
```
